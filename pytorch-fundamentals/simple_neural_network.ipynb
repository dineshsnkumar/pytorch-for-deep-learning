```python
import torch
import torch.nn as nn
import torch.optim as optim
```

### ML Pipeline

### 1. Data Ingestion and Preperation

```python
distances = torch.tensor([[1.0], [2.0], [3.0], [4.0]], dtype=torch.float32)
times = torch.tensor([[6.96], [12.11], [16.77], [22.21]], dtype=torch.float32)
```

#### 2. Model Building

```python
model = nn.Sequential(nn.Linear(1, 1))
```

The goal is to find best weights and bias that fit the linear equation which has a single input (*distances*) and the output (times). For this PyTorch provides a `nn.Sequential(nn.Linear(1, 1))`.

Here `nn.Linear(1,1)` means that it is taking one input(distances) and producing one output(times)

#### 3. Training

```python
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

`nn.MSELoss()` is the Mean Squared Error loss function.

`optim.SGD()` is the Stochastic Gradient Descent optimizer. Which adjusts the weights and bias based on the error.

`lr` is the learning rate. It adjusts how big each adjustment should be

```python
for epoch in range(500):
  # Clear the gradients
  optimizer.zero_grad()
  # Make predictions (Forward pass)
  outputs = model(distances)
  # Calculate the loss
  loss = loss_function(outputs, times)
  # Calculate adjustments (Backward pass)
  loss.backward()
  # Update the model parameters
  optimizer.step()

  if (epoch + 1) % 50 == 0:
    print(f"Epoch {epoch+1} : Loss {loss.item()}")
```

#### Making predictions

```python
distance_to_predict = 7.0

with torch.no_grad():
  new_distance = torch.tensor([distance_to_predict], dtype=torch.float32)
  # Pass. the distance to predict to the model
  prediction = model(new_distance)
  print(f"The predicted time to travel {distance_to_predict} is {prediction.item():.1f}")
```

#### Conclusion

- This model works only when the relationship between the data is linear. For non-linear systems it will fail to fit the data
